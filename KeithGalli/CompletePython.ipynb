```python
# import lib
import pandas as pd
```

```python
# lendo o arquivo CSV
df = pd.read_csv(r'pokemon_data.csv')
df.head()
```

```python
df.info()
```

```python
# Descrição das colunas
print(df.columns)
```

```python
#Filtrando colunas
df[['Name', 'Type 1']]
```

```python
# Filtrando linhas pelas posições
df.iloc[1:4]
```

```python
# Filtro de linha pelo nome
df.loc[df['Name'] == 'Venusaur']
```

```python
# Filtro de linha e coluna
df.loc[df['Name'] == 'Venusaur', 'HP']
```

```python
# Filtro de linha por tipo de coluna
df.loc[df['Type 1'] == 'Fire']
```

```python
# Análise estatisca dos dados
df.describe()
```

```python
# Ordena os valores por Type 1 e HP na ordem ascendente
df.sort_values(['Type 1', 'HP'])
```

```python
# Ordena os valores por Type 1 e HP na ordem descendente
df.sort_values(['Type 1', 'HP'], ascending=False)
```

```python
# Organizando do Type 1 ascendente de HP descendente
df.sort_values(['Type 1', 'HP'], ascending=[1,0])
```

```python
# Organizando do Type 1 descendente de HP ascendente
df.sort_values(['Type 1', 'HP'], ascending=[0,1])
```

```python
# Criando uma nova coluna e somando valores atraves de registros
df['Total'] = df['HP']  + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df
```

```python
# Outra forma de criar coluna e somar registros
df['Total'] = df.iloc[:, 4:10].sum(axis='columns')
df
```

```python
# Salvando o arquivo em csv e xlsx
df.to_csv('modified.csv', index=False, sep=';')
df.to_excel('modified.xlsx', index=False, sheet_name='data')
```

```python
# Filtro de linha baseado em colunas
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)].reset_index(drop=True)
```

```python
# Filtro utilizando parte de uma string
df.loc[df['Name'].str.contains('Mega')]
```

```python
import re
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

```python
df.loc[df['Total'] > 500, ['Generation', 'Legendary']]
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=True)
```

```python
df.groupby(['Type 1']).sum()
```

```python
df.groupby(['Type 1']).count()
```

```python
df['Count'] = 1
df.groupby(['Type 1']).count()['Count']
```
