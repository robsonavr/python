#Análise de dados reembolso senadores 2018

```python
#import bibliotecas
import pandas as pd
```

```python
df_original = pd.read_csv(r'despesa_ceaps_2018.csv', 
                         encoding='latin_1', sep=';', decimal=',', thousands='.',
                         skiprows=1)

display(df_original.head())
```

```python
display(df_original.tail())
```

```python
print(df_original.dtypes)
```

```python
print(df_original.info())
```

```python
print(f'A base tem um total de {df_original.shape[0]} linhas e {df_original.shape[1]} colunas')
```

```python
print(df_original.describe())
```

```python
#Total de reembolso
print(f"Valor total reembolsado é de R$ {df_original['VALOR_REEMBOLSADO'].sum():,.2f}")
```

```python
#Total de recibos por Senador
display(df_original['SENADOR'].value_counts())
```

```python
#Valor total do reembolso por senador
display(df_original.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False))
```

```python
#Os 5 maiores valores de reembolso
display(df_original.nlargest(5, 'VALOR_REEMBOLSADO').T)
```

```python
#Os 5 menores valores de reembolso
display(df_original.nsmallest(5, 'VALOR_REEMBOLSADO').T)
```

```python
#Quantidade de reembolso por tipo de despesa
df_original['TIPO_DESPESA'].value_counts()
```

```python
#Valores totais em relação ao tipo de despesas
display(df_original.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False))
```
