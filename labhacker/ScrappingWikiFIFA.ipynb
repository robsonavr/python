# Análise de dados da copa do mundo na wikipedia

Publico pagante em cada final

https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA

```python
#importacao das bibliotecas

import pandas as pd
import requests as req

url = 'https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA'

res = req.get(url)

table = pd.read_html(res.text)

df = table[11]

# df.columns = df.iloc[0] # define o nome das colunas utilizando a primeira linha

# df.drop(df.index[0], inplace=True) # apaga a primeira linha

df.head()
```

```python
df.info()
```

```python
df['Público pagante'] = df['Público pagante'] * 1000
```

## Construção de graficos

```python
# Importação de bibliotecas graficas

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# exibindo um grafico de linha

ax = sns.pointplot(x='Ano', y='Público pagante', data=df)

plt.xticks(rotation=65)
for x, y in zip(df['Ano'], df['Público pagante']):
#     print(f'Valor de x: {x} e o valor de y: {y}')
    ax.text(x, y-1000, s='{:.0f}'.format(y))
plt.show()
```
