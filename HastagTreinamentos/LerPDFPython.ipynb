# Como Ler Tabelas em PDF Usando o Python [Extrair Tabelas de um Arquivo PDF]

```python
import tabula
import pandas as pd
import datetime
```

```python
#le arquivo pdf em busca de tabelas no arquivo, é possível definir o número da página
#
#parametros para teste:
# guess = False ou lattice = True

gefasoft = tabula.read_pdf('Fgts/extrato_GEFASOFT_BRASIL_LTDA.pdf', pages='all')
asa = tabula.read_pdf('Fgts/extrato_ASA_ROBOTICA_COMERCIO_DE_FERRAGENS_LTDA.pdf', pages='all')
cube = tabula.read_pdf('Fgts/extrato_CUBE_TECNOLOGIA_LTDA.pdf', pages='all')
bystronic = tabula.read_pdf('Fgts/extrato_BYSTRONIC_DO_BRASIL_LTDA.pdf', pages='all')
print([len(gefasoft), len(asa), len(cube), len(bystronic)])
```

```python
#Cria um novo dataframe
fgts = pd.DataFrame(columns = ['DATA', 'LANÇAMENTO', 'VALOR', 'TOTAL'])
fgts
```

```python
#Adiciona a tabela gefasoft
indice = []
for n_tabela in range(len(gefasoft)):
    if 'LANÇAMENTO' in gefasoft[n_tabela].columns:
        indice.append(n_tabela)
print(indice)

frame = []
for i in indice:
    frame.append(gefasoft[i].loc[gefasoft[i]['LANÇAMENTO'].str.contains('CREDITO DE JAM')])

gefasoft = pd.concat(frame, ignore_index=True)
```

```python
#Adiciona a tabela asa robotica
indice = []
for n_tabela in range(len(asa)):
    if 'LANÇAMENTO' in asa[n_tabela].columns:
        indice.append(n_tabela)
print(indice)

frame = []
for i in indice:
    frame.append(asa[i].loc[asa[i]['LANÇAMENTO'].str.contains('CREDITO DE JAM')])

asa = pd.concat(frame, ignore_index=True)
```

```python
#Adiciona a tabela cube
indice = []
for n_tabela in range(len(cube)):
    if 'LANÇAMENTO' in cube[n_tabela].columns:
        indice.append(n_tabela)
print(indice)

frame = []
for i in indice:
    frame.append(cube[i].loc[cube[i]['LANÇAMENTO'].str.contains('CREDITO DE JAM')])

cube = pd.concat(frame, ignore_index=True)
```

```python
#Adiciona a tabela bystronic
indice = []
for n_tabela in range(len(bystronic)):
    if 'LANÇAMENTO' in bystronic[n_tabela].columns:
        indice.append(n_tabela)
print(indice)

frame = []
for i in indice:
    frame.append(bystronic[i].loc[bystronic[i]['LANÇAMENTO'].str.contains('CREDITO DE JAM')])

bystronic = pd.concat(frame, ignore_index=True)
```

```python
fgts = pd.concat([gefasoft, asa, cube, bystronic])
```

```python
fgts['DATA'] = pd.to_datetime(fgts['DATA'], format = '%d/%m/%Y')
```

```python
fgts.info()
```

```python
fgts['LANÇAMENTO'] = fgts['LANÇAMENTO'].str.split(' ', expand=True)[3]
```

```python
fgts['LANÇAMENTO'] = fgts['LANÇAMENTO'].str.replace(',', '.').astype('float64')
```

```python
fgts.info()
```

```python
fgts['VALOR'] = fgts['VALOR'].str.replace(',', '.')
fgts['VALOR'] = fgts['VALOR'].str.replace('[R\$]', '').astype('float64')
```

```python
fgts.info()
```

```python
fgts['base'] = fgts['VALOR'] / fgts['LANÇAMENTO']
fgts
```
