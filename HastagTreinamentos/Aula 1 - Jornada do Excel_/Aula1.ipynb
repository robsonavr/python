```python
# importando libs
import pandas as pd
import datetime as dt
```

```python
df = pd.read_excel(r'Dashboard da Jornada do Excel - Aula 1.xlsx', sheet_name='ZERO')
```

```python
# corrigindo a data

df['DATA VENDA'] = pd.TimedeltaIndex(df['DATA '], unit='D') + dt.datetime(1899,12,30)
```

```python
df.info()
```

```python
df.tail()
```

```python
# criando uma tabela calendario

def create_calendar(start='2000-01-01', end='2022-12-31'):
    df = pd.DataFrame({"Date": pd.date_range(start, end)})
    df["Day"] = df.Date.dt.day
    df["Week"] = df.Date.dt.isocalendar().week
    df["Week_day"] = df.Date.dt.weekday
    df["Month"] = df.Date.dt.month
    df["Month_name"] = df.Date.dt.month_name()
    df["Month_abbr"] = df.Date.dt.month_name().str[:3]
    df["Quarter"] = df.Date.dt.quarter
    df["Year"] = df.Date.dt.year
    df["Year_half"] = (df.Quarter + 1) // 2
    return df

dCalendario = create_calendar(df['DATA VENDA'].min(), df['DATA VENDA'].max())

dCalendario
```

```python
# relacionamento tabela fato com tabela calendario

df_temporal = df.set_index('DATA VENDA').join(dCalendario.set_index('Date'), how='left')
df_temporal
```

```python
#Quantidade de produtos na lista por tipo
df['PRODUTO'].value_counts().sort_index()
```

```python
# Valor total por produto
df.groupby('PRODUTO')['TOTAL'].sum().reset_index().style.format({'TOTAL': 'R$ {:,.2f}'})
#df.groupby('PRODUTO')['TOTAL'].sum().map('R$ {:,.2f}'.format)
```

```python
# Valor Totol
print(f"Valor total em produtos R$ {df['TOTAL'].sum():,.2f}")
```

```python
# valor total dos produtodos pelo ano
df_temporal.groupby(['PRODUTO', 'Year'], as_index=False)['TOTAL'].sum().pivot('PRODUTO', 'Year').style.format('R$ {:,.2f}')
```

```python
#Valor total de produtos pelo ano e tipo de pagamento
for pagamento in df_temporal['FORMA PGTO'].unique():
    df_temp = df_temporal[df_temporal['FORMA PGTO'] == pagamento]
    print(f'\n\nTipo de pagamento: {pagamento}')
    display(df_temp.groupby(['PRODUTO', 'Year'], as_index=False)['TOTAL'].sum().pivot('PRODUTO', 'Year').style.format('R$ {:,.2f}'))
    
```

```python
# valor total de venda por mês durante todos os anos
df_temporal.groupby(['Month', 'Month_name'], as_index=False)['TOTAL'].sum().drop(columns='Month').style.format({'TOTAL':'R$ {:,.2f}'})
```

```python
# valor total de venda por mês e por anos
df_temporal.groupby(['Month', 'Year'], as_index=False)['TOTAL'].sum().pivot('Month', 'Year')
```
