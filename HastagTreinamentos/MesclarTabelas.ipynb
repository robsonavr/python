```python
import pandas as pd
```

## Caso 1

```python
df_fvendas = pd.read_excel('Exemplo 1.xlsx', sheet_name='BaseVendas')
df_dlojas = pd.read_excel('Exemplo 1.xlsx', sheet_name='CadastroLojas')
df_dprodutos = pd.read_excel('Exemplo 1.xlsx', sheet_name='CadastroProdutos')
```

```python
display(df_fvendas.head(1))
display(df_dlojas.head(1))
display(df_dprodutos.head(1))
```

```python
df_fvendas.info()
df_dlojas.info()
df_dprodutos.info()
```

```python
df = pd.merge(df_fvendas, df_dlojas, on='Loja', how='left')
```

```python
df.loc[df['Estado'] == 'PR'].count()
```

```python
df = pd.merge(df, df_dprodutos, on='SKU', how='left')
```
