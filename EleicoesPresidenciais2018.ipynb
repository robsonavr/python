```python
#Carregando os arquivos da eleição do segundo turno para presidente

df = pd.read_csv(r'./votacao_candidato_munzona_2018/votacao_candidato_munzona_2018_BR.csv', 
                 encoding='latin1', 
                 sep=';')
df.head().T
```

```python
#Armazenando somente dados do segundo turno, somente nos municipios nacionais
df = df[(df['NR_TURNO'] == 2) & (df['SG_UF'] != 'ZZ')]
```

```python
#Quantidade de votos por candidato
df.groupby('NM_CANDIDATO')['QT_VOTOS_NOMINAIS'].sum()
```

```python
#Eliminar colunas que não serão utilizadas
df_votacao = df[['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 'NM_CANDIDATO',
        'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS']]
df_votacao.head()
```

```python
#Agregação de municipios por candidatos
df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 
                    'NM_URNA_CANDIDATO']).sum()
```

```python
#Reindexação do indice do dataframe
df_votacao = df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO', 
                    'NM_URNA_CANDIDATO']).sum().reset_index()
df_votacao.sample(4)
```

```python
#Ordenar os valores e remover os municiopios duplicados mantendo o primeiro

#Ordenando valores
df_votacao.sort_values('QT_VOTOS_NOMINAIS', ascending=False, inplace=True)

#Remove duplicados e mantem o primeiro registro
df_votacao.drop_duplicates('CD_MUNICIPIO', keep='first', inplace=True)

df_votacao.head()
```

```python
#Criando uma coluna com o vencedor do municipio True=Bolsonaro e False=outro
df_votacao['RESULTADO'] = df_votacao['NR_CANDIDATO'].apply(lambda x : x == 17)
```

```python
#Tabela auxiliar para relacionamento das tabelas do IBGE e TSE
url = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'

df_auxiliar = pd.read_csv(url)

df_auxiliar.head()
```

```python
#Reindexacao das tabelas
#df_votacao.set_index('CD_MUNICIPIO', inplace=True)
#df_auxiliar.set_index('COD_TSE', inplace=True)
```

```python
#Reduzir a tabela auxiliar
df_auxiliar = df_auxiliar[['GEOCOD_IBGE', 'COD_TSE']]
```

```python
#Como os registros são unicos a quantidade de registros das tabelas tem ser iguais / relacao 1 para 1

df_votacao.info()
```

```python
df_auxiliar.info()
```

```python
df_auxiliar.head()
```

```python
#unindo tabela auxiliar na tabela votacao usando join
df_unificada = df_votacao.set_index('CD_MUNICIPIO').join(df_auxiliar.set_index('COD_TSE'))
df_unificada.head()
```

```python
#Executa o relacionamento com a tabela mapa
df_mapa.info()
df_mapa.head(1)
```

```python
#Unindo a tabela votacao a tabela mapa
df_unificada['GEOCOD_IBGE'] = df_unificada['GEOCOD_IBGE'].astype(str)
df_final = df_mapa.set_index('CD_MUN').join(df_unificada.set_index('GEOCOD_IBGE'))
# gdf = geopandas.GeoDataFrame(
#     df, geometry=geopandas.points_from_xy(df.Longitude, df.Latitude))
df_final = gpd.GeoDataFrame(df_final)
df_final.head()
```

## Desenhar o Mapa

```python
#Codigo de cores False/0 = red e True/1 = green
from matplotlib.colors import ListedColormap

cmap = ListedColormap(['red', 'green']) # define as cores do mapa

#Desenha o mapa
fig, ax = plt.subplots(figsize=(12,12)) 

figura = df_final.plot(column='RESULTADO', cmap=cmap, legend=True, linewidth=0.05, ax=ax, edgecolor='gray')

ax.set_axis_off() # esconde os eixos

ax.set_title('Resultado do Segundo Turno 2018', fontdict={'fontsize':25, 'fontweight':3}) # Titulo

ax.annotate('Fontes IBGE, TSE e Estadão de São Paulo',
           xy=(0.1, 0.08), xycoords='figure fraction', horizontalalignment='left',
           verticalalignment='top', fontsize=10, color='#555555') # Rodape

legenda = figura.get_legend()
print(legenda)
# legenda.get_texts()[0].set_text('Fernando Haddad')
# legenda.get_texts()[1].set_text('Jair Bolsonaro')

plt.show()
```
