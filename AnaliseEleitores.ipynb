```python
#importacao das bibliotecas
import pandas as pd
```

```python
#Carregando o arquivo do dataframe
df = pd.read_csv(r'eleitorado_municipio_2020.csv', sep=';', encoding='latin_1')
```

```python
#Maior quantidade de eleitores deficientes
df.nlargest(5, 'QTD_ELEITORES_DEFICIENTE')
```

```python
#Quantidade de generos 
print('Quantidade de eleitoras', df['QTD_ELEITORES_FEMININO'].sum())
print('Quantidade de eleitores', df['QTD_ELEITORES_MASCULINO'].sum())
print('Quantidade de não informados', df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
#Calculo do percentual
total_eleitores = df['QTD_ELEITORES'].sum()
total_feminino = df['QTD_ELEITORES_FEMININO'].sum()
total_masculino = df['QTD_ELEITORES_MASCULINO'].sum()
total_naoinformado = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
print(f'Total eleitoras: {(total_feminino / total_eleitores):.2%}')
print(f'Total eleitores: {(total_masculino / total_eleitores):.2%}')
print(f'Total não informado: {(total_naoinformado / total_eleitores):.2%}')
```

```python
#Municipio onde a quantidade de eleitores masculino é maior que o feminino
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
#Relacao feminino / Masculino
df['RELACAO F-M'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```python
#Municipios com a maior relacao feminino / masculino
df.nlargest(5, 'RELACAO F-M').T
```

```python
#Municipios com a menor relacao feminino / masculino
df.nsmallest(5, 'RELACAO F-M').T
```

```python
df['CD_PAIS'].value_counts()
```

```python
#Graficos da relacao masculino / feminino
# mostra o grafico apos comando
%matplotlib inline

df['RELACAO F-M'].plot.hist(bins=100)
```

```python
#importar a biblioteca grafica
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.displot(df['RELACAO F-M'], bins=100, color='red', kde=True)

plt.title('Relação entre eleitores feminino e masculino', fontsize=18)
plt.xlabel('Eleitores / Eleitoras', fontsize=14)
plt.ylabel('Frequencia', fontsize=14)
plt.axvline(1.0, color='black', linestyle='--')
```

```python
#Distribuição feminino / masculino em pontos
sns.stripplot(data=df, x='NM_REGIAO', y='RELACAO F-M', jitter=0.3, size=3.0)
plt.axhline(1.0, color='black', linestyle='--')
```
